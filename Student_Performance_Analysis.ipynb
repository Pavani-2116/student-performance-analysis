**Data Preparation**

```python
#Create average score
df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3
df
```

Explanation:

Creates a new column called average_score which represents the average exam score of each student.

**Exploratory Data Analysis (EDA)**

```python
#Distribution of student scores
plt.hist(df["average_score"], bins=20)
plt.title("Distribution of Student Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

Explanation:

Shows how student scores are distributed across the dataset.
This histogram shows the distribution of students' average scores. It helps understand how scores are spread and where most students' performance lies.

```python
#Impact of Test Preparation
sns.barplot(x="test preparation course", y="average_score", data=df)

plt.title("Impact of Test Preparation on Scores")
plt.show()
```

Explanation:

Compares average scores between students who:

This chart compares average scores between students who completed the test preparation course and those who did not, helping identify its impact on performance.

```python
#Subject Performance Comparison
df[["math score","reading score","writing score"]].mean().plot(kind="bar")

plt.title("Average Score by Subject")
plt.ylabel("Score")
plt.show()
```

Explanation

Compares average scores across subjects:

Math

Reading

Writing
This bar chart compares the average scores across math, reading, and writing subjects to understand which subject students perform better in.

```python
#Correlation Between Subjects
sns.heatmap(
    df[["math score","reading score","writing score"]].corr(),
    annot=True
)

plt.title("Correlation Between Subjects")
plt.show()
```

Explanation:

Shows how strongly the subjects are related.

Example:

reading ↔ writing often have strong correlation.
This heatmap shows the correlation between subject scores, helping identify how strongly the subjects are related to each other.

```python
#Gender vs Performance
sns.boxplot(x="gender", y="average_score", data=df)

plt.title("Average Score by Gender")
plt.show()
```

Explanation:

Shows how average scores vary between male and female students.
This box plot compares the distribution of average scores between male and female students, highlighting differences in performance and score variability.



```python
#Parental Education Impact
sns.barplot(x="parental level of education", y="average_score", data=df)

plt.xticks(rotation=45)
plt.title("Average Score by Parental Education")
plt.show()
```

Explanation:

Shows how parental education levels influence student academic performance.

**Key Insights**


• Students who completed the test preparation course scored higher on average.

• Reading and writing scores show a strong positive correlation.

• Math scores are slightly lower compared to reading and writing scores.

• Most students scored between 60 and 80 average marks.


**Conclusion**


This analysis explored factors influencing student academic performance.

The results show that test preparation programs can improve student scores and that reading and writing abilities are strongly related.
